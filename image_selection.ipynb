```python
import glob
import shutil
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
#xml에서 원하는 tag 추출
def img_tag_list(xml,tag):
    
    imgs=[]
    f=open(xml,'rt',encoding='UTF-8')
    soup=BeautifulSoup(f,"lxml")
    object_list=soup.find_all("box",{"label":tag})
   
    for i in range(len(object_list)):
        #path=xml[:60]+(object_list[i].parent)["name"]
        imgs.append(object_list[i].parent)
        #imgs.append(path)
        
    f.close()
            
    imgs=list(set(imgs)) #set으로 중복 제거후 list로 다시 형 변환

    
    return imgs
    
```

```python
#xml 주소 리스트 반환
def xml_list(start,end): #start~end 폴더 번호
    s=r'C:/Users/mike5/Desktop/walking_video/bounding_box/Bbox_'
    xml_list=[]
    for i in range(start,end+1):
        path=s+str(i).zfill(4)+"/*.xml"
        file=glob.glob(path)
        xml_list.append(file[0]) #경로당 xml파일 1개이기때문
    
    return xml_list
```

```python
#create annotation
def write_xml(image_list,new_xml_path):
    
    new_xml_file=open(new_xml_path,'a',encoding='UTF-8')
    for i in range(len(image_list)):
        new_xml_file.write(str(image_list[i]))
        new_xml_file.write('\n')
            
    new_xml_file.close()
```

```python
#test
xml=xml_list(1,1)
img_list=img_tag_list(xml,"bicycle")
print(img_list)
```

```python
path='C:/Users/mike5/Desktop/walking_video/bicycle_images'
test_path='C:/Users/mike5/Desktop/walking_video/bicycle_test_images'
start=1
end=350

xml=xml_list(start,end)

for i in tqdm(len(xml)):
    image_list=img_tag_list(xml[i],"bicycle")
    write_xml(image_list,path+'/annotation.xml')
    for j in range(len(image_list)):
        name=image_list[j]["name"]
        org_path='C:/Users/mike5/Desktop/walking_video/bounding_box/Bbox_'
        org_path=org_path+str(i+start).zfill(4)+'/'+name
        
        shutil.copy2(org_path,path+'/'+name)    
```
